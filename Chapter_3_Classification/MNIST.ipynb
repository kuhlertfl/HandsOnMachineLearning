# MNIST

```python
from sklearn.datasets import fetch_openml
from sympy.utilities.iterables import random_derangement

# fetch_openml returns the inputs as Pandas Dataframe and the labels as a Pandas Series
# Here the dataset contains images so data frames are not good so we chose as_frame = False
mnist_data = fetch_openml('mnist_784', as_frame=False) # as_frame imports it directly as panda dataset


```

```python
print(type(mnist_data))
print(f"INFORMATION ABOUT THE DATA SET:\n{mnist_data.DESCR}")
print(f"THE DATA:\n{mnist_data.data}")
print(f"THE LABELS:\n{mnist_data.target}")
```

```python
# Load data in variables
X,y = mnist_data.data, mnist_data.target
```

```python
import matplotlib.pyplot as plt

# The image has 784 features because its 28x28 pixels (each from 0-255 intensity)
# To see the image we reshape the array into the picture array and use matplotlib to plot it
def plot_digit(image_data):
    image = image_data.reshape(28,28)
    plt.imshow(image, cmap='binary')
    plt.axis("off")
    #print(image)

some_digit = X[10]
plot_digit(some_digit)
plt.show()
print(f"Label: {y[10]}")
```

```python
# Split the data set into train and test set
# the MNIST dataset is already splitted so we can just
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
# else we would have to make sure that the labels are equally divided int otest and training set >> see chapter 2
```

## Training a Binary Classifier

```python
# Start with a binary classifier
y_train_5 = (y_train == '5')
y_test_5 = (y_test == '5')


from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
sgd_clf.predict([some_digit])
```
